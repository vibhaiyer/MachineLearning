```python
import pandas as pd
import numpy as np 
import statistics
import math
from numpy import genfromtxt
dataset = pd.read_csv("expectation.csv", header = 0)
print(dataset)

#Assumption
#There are 2 clusters
Mean_1 = 111
SD_1 = 20

Mean_2 = 160
SD_2 = 20
```

```python
#Expectation Step

list_Z1 =[]
list_Z2 = []
for index, row in dataset.iterrows():
    print("For Index", index+1)
    x = row[1]
    a1 = SD_1 * math.sqrt(2*math.pi)
    b1 = -pow((x-Mean_1),2)
    c1 = 2*SD_1*SD_1
    f1 = (1/a1)*(math.exp(b1/c1))
    print("\tFor cluster 1 f(x) is", f1)
    x = row[1]
    a2 = SD_1 * math.sqrt(2*math.pi)
    b2 = -pow((x-Mean_2),2)
    c2 = 2*SD_2*SD_2
    f2 = (1/a2)*(math.exp(b2/c2))
    print("\tFor cluster 2 f(x) is", f2)
    z1 = f1/(f1+f2)
    z2 = f2/(f1+f2)
    print("\tz1 is", z1)
    list_Z1.append(z1)
    print("\tz2 is", z2)
    list_Z2.append(z2)
```

```python
#Maximization Step
Mean1_num = 0
den1 = 0
SD1_num = 0
Mean2_num = 0
den2 = 0
SD2_num = 0
for index, row in dataset.iterrows():
    x = row[1]
    #Cluster1
#    print("For Cluster 1")
    Mean1_num = Mean1_num + list_Z1[index]*x
    den1 = den1 + list_Z1[index]
    SD1_num = SD1_num + (list_Z1[index] * pow((x-Mean_1),2))
    #Cluster2
    Mean2_num = Mean2_num + list_Z2[index]*x
    den2 = den2 + list_Z2[index]
    SD2_num = SD2_num + (list_Z2[index] * pow((x-Mean_2),2))   

new_Mean_1 = Mean1_num/den1
new_SD_1 = math.sqrt(SD1_num/den1)
new_Mean_2 = Mean2_num/den2
new_SD_2 = math.sqrt(SD2_num/den2)

print("New MEan1", new_Mean_1)
print("New SD1", new_SD_1)
print("New Mean2",new_Mean_2)
print("New SD2",new_SD_2)
```
